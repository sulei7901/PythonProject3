```
#!pip install pandas
#!pip install matplotlib
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Dataset columns:**

* extract_date = Date of extraction from live disease surveillance database

* specimen_date = Date of specimen collection, equivalent to diagnosis date

* Number_tested = Count of NYC residents newly tested for SARS-CoV-2

* Number_hospitalized = Count of patients tested who were confirmed to be COVID-19 cases

* Number_deaths = Count of confirmed COVID-19 cases among patients who died

```
#Import data
dfcsv = df = pd.read_csv('https://raw.githubusercontent.com/sulei7901/datasets/master/COVID-19_Outcomes_by_Testing_Cohorts__Cases__Hospitalizations__and_Deaths.csv')
df
df.dtypes

#Change the specimen_date column into datetime format
df['specimen_date'] = pd.to_datetime(df.specimen_date) 
df.dtypes

#Filter for extract date 08/02/2020, 160 rows x 6 columns
df.set_index('extract_date', inplace=True)
df
df2=df.loc['08/02/2020']
df2
df2.describe()

#Sort by specimen_date
df2=df2.sort_values(by='specimen_date')
df2

#Add Month column
df2['Month'] = pd.DatetimeIndex(df2['specimen_date']).month

#Show all rows
#pd.set_option('display.max_rows', None)

#Rename columns
df2=df2.rename(columns={'specimen_date':'Specimen Date',
                        'Number_tested':'Number Tested', 
                        'Number_confirmed':'Number Confirmed',
                        'Number_hospitalized':'Number Hospitalized',
                        'Number_deaths':'Number Deaths'})
df2
```

## **Tested and Confirmed**

```
# Plot showing number tested and number confirmed by specimen_date
df2[-150:].plot(x='Specimen Date', y=['Number Tested', 'Number Confirmed'], figsize=(20,9), grid=True)
plt.title('DAILY NUMBER TESTED AND NUMBER CONFIRMED')
plt.show
```

```
# Plot showing total number tested and number confirmed by month
data1=['Number Tested', 'Number Confirmed']
df2.groupby(['Month'])[data1].sum().plot(kind='bar', figsize=(20,9))
plt.title('MONTHLY TOTAL NUMBER TESTED AND NUMBER CONFIRMED')
plt.show
```

## **Confirmed and Hospitalized**

```
# Plot showing daily number confirmed and number hospitalized
df2[-150:].plot(x='Specimen Date', y=['Number Confirmed', 'Number Hospitalized'], figsize=(20,9), grid=True)
plt.title('DAILY NUMBER CONFIRMED AND NUMBER HOSPITALIZED')
plt.show
```

```
# Plot showing total number confirmed and number hospitalized by month
data2=['Number Confirmed', 'Number Hospitalized']
df2.groupby(['Month'])[data2].sum().plot(kind='bar', figsize=(20,9))
plt.title('MONTHLY TOTAL CONFIRMED AND TOTAL HOSPITALIZED')
plt.show
```

##**Hospitalized and Deaths**




```
# Plot showing daily number hospitalized and number deaths
df2[-150:].plot(x='Specimen Date', y=['Number Hospitalized', 'Number Deaths'], figsize=(20,9), grid=True)
plt.title('DAILY NUMBER HOSPITALIZED AND NUMBER DEATHS')
plt.show
```

```
# Plot showing total number hospitalized and number deaths by month
data3=['Number Hospitalized', 'Number Deaths']
df2.groupby(['Month'])[data3].sum().plot(kind='bar', figsize=(20,9))
plt.title('MONTHLY TOTAL HOSPITALIZED AND NUMBER DEATHS')
plt.show
```

## **All Data**

```
# Plot showing all daily totals
df2[-150:].plot(x='Specimen Date', y=['Number Tested', 'Number Confirmed', 'Number Hospitalized', 'Number Deaths'], figsize=(20,9), grid=True)
plt.title('DAILY TOTALS')
plt.show
```

```
# Plot showing all data monthly totals
data4=['Number Tested', 'Number Confirmed', 'Number Hospitalized', 'Number Deaths']
df2.groupby(['Month'])[data4].sum().plot(kind='bar', figsize=(20,9))
plt.title('MONTHLY TOTALS')
plt.show
```
