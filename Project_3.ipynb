```
#!pip install pandas
#!pip install matplotlib
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Dataset columns:**
  
* extract_date = Date of extraction from live disease surveillance database

* specimen_date = Date of specimen collection, equivalent to diagnosis date

* Number_tested = Count of NYC residents newly tested for SARS-CoV-2

* Number_hospitalized = Count of patients tested who were confirmed to be COVID-19 cases

* Number_deaths = Count of confirmed COVID-19 cases among patients who died

```
dfcsv = df = pd.read_csv('https://raw.githubusercontent.com/sulei7901/datasets/master/COVID-19_Outcomes_by_Testing_Cohorts__Cases__Hospitalizations__and_Deaths.csv')
df
df.dtypes

#Filter for extract date 08/02/20, 160 rows x 6 columns
extdate = '08/02/2020'
df2 = df[df['extract_date'] == extdate]  
df2
df2.describe()

#Change the specimen_date column into datetime format
df2['specimen_date'] = pd.to_datetime(df.specimen_date) 
df2.dtypes

#Add month column
df2['month'] = pd.DatetimeIndex(df2['specimen_date']).month

# Show all rows
pd.set_option('display.max_rows', None)

#Sort by specimen _date
df2.sort_values(by='specimen_date')
```

```
# Plot number tested and number confirmed by specimen_date
df2[-150:].plot(x='specimen_date', y=['Number_tested', 'Number_confirmed'], figsize=(20,9), grid=True)
plt.title('DAILY NUMBER TESTED AND NUMBER CONFIRMED')
plt.show
```

```
# Groupby
print('Data Extraction as of 08/02/2020')
print('TOTALS BY MONTH')
df2.groupby(['month']).sum()
```

```
# Plot showing total number tested and number confirmed by month
df2.groupby(['month'])['Number_tested', 'Number_confirmed'].sum().plot(kind='bar', figsize=(20,9))
plt.title('MONTHLY TOTAL NUMBER TESTED AND NUMBER CONFIRMED')
plt.show
```

```
# Plot showing all data
df2[-150:].plot(x='specimen_date', y=['Number_tested', 'Number_confirmed', 'Number_hospitalized', 'Number_deaths'], figsize=(20,9), grid=True)
plt.title('DAILY TOTALS')
plt.show
```

```
# Plot showing daily number hospitalized and number deaths
df2[-150:].plot(x='specimen_date', y=['Number_hospitalized', 'Number_deaths'], figsize=(20,9), grid=True)
plt.title('DAILY NUMBER HOSPITALIZED AND NUMBER DEATHS')
plt.show
```

```
# Plot showing total number hospitalized and number deaths by month
df2.groupby(['month'])['Number_hospitalized', 'Number_deaths'].sum().plot(kind='bar', figsize=(20,9))
plt.title('MONTHLY TOTAL HOSPITALIZED AND NUMBER DEATHS')
plt.show
```

```
# Plot showing daily number confirmed and number hospitalized
df2[-150:].plot(x='specimen_date', y=['Number_confirmed', 'Number_hospitalized'], figsize=(20,9), grid=True)
plt.title('DAILY NUMBER CONFIRMED AND NUMBER HOSPITALIZED')
plt.show
```

```
# Plot showing total number confirmed and number hospitalized by month
df2.groupby(['month'])['Number_confirmed', 'Number_hospitalized'].sum().plot(kind='bar', figsize=(20,9))
plt.title('MONTHLY TOTAL CONFIRMED AND TOTAL HOSPITALIZED')
plt.show
```
